```python
from datetime import datetime
```

```python
def get_input():
    return input("Enter command: ").strip().lower()
```

```python
def decide(query):
    if "calculate" in query or any(op in query for op in "+-*/"):
        return "calc"
    if "date" in query:
        return "date"
    if "hello" in query or "hi" in query:
        return "greet"
    return "unknown"
```

```python
def act(intent, query):
    if intent == "calc":
        expr = query.replace("calculate", "").strip()
        try:
            return eval(expr)
        except:
            return "Invalid expression"

    if intent == "date":
        return datetime.now().strftime("%Y-%m-%d")

    if intent == "greet":
        return "Hello"

    return "Unknown command"
```

```python
while True:
    q = get_input()
    if q in {"exit", "quit"}:
        break
    print("Result:", act(decide(q), q))
```
