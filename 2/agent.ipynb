```python
from tools import TOOLS
```

```python
def decide_tool(query: str):
    q = query.lower()

    if any(k in q for k in ["calculate", "compute", "+", "-", "*", "/"]):
        expr = q.replace("calculate", "").replace("compute", "")
        expr = expr.replace("plus", "+").replace("minus", "-")
        expr = expr.replace("times", "*").replace("divided by", "/")
        return "calculator", expr.strip()

    if "weather" in q:
        words = q.split()
        city = words[-1] if len(words) > 1 else "delhi"
        return "weather", city

    if any(k in q for k in ["summarize", "summary"]):
        text = q.replace("summarize", "").replace("summary", "")
        return "summarize", text.strip()

    if any(k in q for k in ["hello", "hi", "hey"]):
        return "greet", ""

    if any(k in q for k in ["date", "today"]):
        return "date", ""

    return None, None
```

```python
def run_agent(query: str):
    tool, arg = decide_tool(query)

    if tool not in TOOLS:
        return "No suitable tool found"

    print(f"Tool: {tool} | Arg: {arg}")
    return TOOLS[tool](arg)
```

```python
while True:
    q = input("Enter command: ").strip()

    if q in {"exit", "quit"}:
        break

    print("Result:", run_agent(q))
```
